# lenguaje python, crear la clase "animal" con los atributos "nombre", "raza" crear el conductor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos


```python
# Definición de la clase Animal
class Animal:
    # El método __init__ es el "constructor".
    # Se ejecuta automáticamente al crear un nuevo objeto.
    def __init__(self, nombre, raza):
        self.nombre = nombre  # Atributo: nombre del animal
        self.raza = raza      # Atributo: raza del animal

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} (raza {self.raza}) dice: ¡Guau, guau!🤖 ")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre} está corriendo a toda velocidad.😈 ")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 😛 ")

# --- Creación de Objetos (Instancias) ---

# Creamos el objeto 'pitbull' pasando los datos al constructor
pitbull = Animal("Thor", "Pitbull")

# Creamos el objeto 'chihuahua' pasando sus propios datos
chihuahua = Animal("Pipo", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "Chafa")

# --- Ejecución de Funciones ---

# Usando el objeto pitbull
pitbull.ladrar()
pitbull.correr()

print("-" * 30) # Línea divisoria en la consola

# Usando el objeto chihuahua
chihuahua.ladrar()
chihuahua.correr()
chihuahua.morder()

print("-" * 30) # Línea divisoria en la consola

# Usando el objeto perro callejero
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()
print("programa creado por Axel Guzman NL = 23")
```
